```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

img = np.random.randint(0, 256, (8, 8, 3))
```

```python
plt.imshow(img)
```

```python
upscaled = np.kron(img, np.ones((30, 30, 1)))
Image.fromarray(upscaled.astype(np.uint8)).save('upscaled.png')
```

```python
imported_img = Image.open('./upscaled.png')
imported_img = imported_img.resize((8, 8))
imported_img = np.array(imported_img)
plt.imshow(imported_img)
```

```python
import numpy as np

if upscaled.shape[0] != upscaled.shape[1]:
    print('not valid image')
else:
    size = upscaled.shape[0]
    upscale_factor = upscaled.shape[0] // 8

    rows, cols = np.meshgrid(np.arange(0, size, upscale_factor), np.arange(0, size, upscale_factor))
    pixel_locs = np.column_stack((rows.ravel(), cols.ravel()))

    orig_img = np.zeros((8, 8, 3))
    for i, j in pixel_locs:
        orig_img[i//upscale_factor, j//upscale_factor] = upscaled[i, j]
```
